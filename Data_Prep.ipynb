```python
df = df[df['product_detail'].isin(old_to_new_mapping.keys())].copy()

df['product_detail'] = df['product_detail'].map(old_to_new_mapping)

df['product_id'] = df['product_detail'].apply(lambda x: master_nusantara[x]['new_id'])
df['product_category'] = df['product_detail'].apply(lambda x: master_nusantara[x]['category'])
df['product_type'] = df['product_detail'].apply(lambda x: master_nusantara[x]['type'])
df['unit_price'] = df['product_detail'].apply(lambda x: master_nusantara[x]['price'])

df.to_excel('Dataset/Coffee_Shop_Sales_Nusantara_Clean.xlsx', index=False)
print("Data berhasil di-mapping dan diekspor ke folder Dataset.")
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['transaction_time'] = df['transaction_time'].astype(str)

df['transaction_datetime'] = pd.to_datetime(df['transaction_date'].dt.strftime('%Y-%m-%d') + ' ' + df['transaction_time'])

df['bulan'] = df['transaction_datetime'].dt.month
df['hari'] = df['transaction_datetime'].dt.dayofweek
df['is_weekend'] = df['hari'].apply(lambda x: 1 if x >= 5 else 0)
df['jam'] = df['transaction_datetime'].dt.hour

df['total_penjualan'] = df['unit_price'] * df['transaction_qty']

df[['transaction_datetime', 'bulan', 'hari', 'is_weekend', 'jam', 'transaction_qty']].head()
```

```python
kolom_dihapus = ['transaction_id', 'transaction_date', 'transaction_time', 'product_detail', 'product_id']
df_ml = df.drop(columns=kolom_dihapus)

df_ml = pd.get_dummies(df_ml, columns=['store_location', 'product_category', 'product_type'], dtype=int)

print(f"Jumlah kolom setelah feature engineering & encoding: {df_ml.shape[1]}")
```

```python
df_ml = df_ml.sort_values('transaction_datetime').reset_index(drop=True)

X = df_ml.drop(columns=['transaction_qty', 'total_penjualan'])
y = df_ml['transaction_qty']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print("===== DATA SPLITTING FORECASTING SELESAI =====")
print(f"Batas Waktu Train: Sampai {X_train['transaction_datetime'].max()}")
print(f"Batas Waktu Test : Mulai {X_test['transaction_datetime'].min()}")
```

```python
X_train.to_csv('Dataset/X_train.csv', index=False)
X_test.to_csv('Dataset/X_test.csv', index=False)
y_train.to_csv('Dataset/y_train.csv', index=False)
y_test.to_csv('Dataset/y_test.csv', index=False)

print("Semua dataset Forecasting berhasil diekspor ke folder 'Dataset/'!")
```
